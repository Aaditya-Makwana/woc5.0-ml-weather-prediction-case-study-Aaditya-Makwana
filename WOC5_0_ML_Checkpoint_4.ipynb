#Matplotlib

```python
from matplotlib import pyplot as plt
plt.plot(df.car_ID, df.price)
plt.title('Car ID vs Price')
plt.xlabel('Car ID')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(df.car_ID, df.citympg, label = 'City', marker = '.', color = 'b')
plt.plot(df.car_ID, df.highwaympg, label = 'Highway', marker = '.', color = 'k')
plt.xlabel('Car ID')
plt.ylabel('Miles per Gallon')
plt.title('City Mileage vs Highway Mileage')
plt.legend()
plt.grid(True)
plt.show()
```
